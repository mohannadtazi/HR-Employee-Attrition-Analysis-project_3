### encode categorial column `Attrition` and drop other column not used

```python
df.head(1)
```

```python
df_num = df.loc[:,:]
dict_num1={'Yes':1 , 'No':0}
df_num['Attrition']=df_num['Attrition'].replace(dict_num1)
df_num = df_num.drop(['BusinessTravel', 'Department' , 'EducationField','Gender','JobRole','MaritalStatus','Over18','OverTime'] , axis=1)
```

### correlation between numeric values

```python
# Calculate correlation matrix
correlation_matrix = df_num.corr()

# Plot correlation heatmap
plt.figure(figsize=(20, 18))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
